Un nudo de *Lissajous* K en $\mathbb{R}^{3}$ tiene la parametrizaci√≥n, $K(t)= (x(t),y(t),z(t))$, donde 

\begin{split}
x(t) =& \cos(n_{x}t+ \phi_{x})\\
y(t) = & \cos(n_{y}t+\phi_{y})\\
z(t) =& \cos(n_{z}t+\phi_{z})
\end{split}
donde $0 \le t \le 2\pi$, $n_{x},n_{y},n_{z} \in \mathbb{N}$ y $\phi_{x},\phi_{y},\phi{z} \in \mathbb{R}$.

Los nudos *Lissajous* es un caso particular del caso general de los nudos de *Fourier*.

Un nudo con coeficientes de Fourier-$(i,j,k)$ est√° determiniado por las coordenadas   
$$x_{t}=A_{x,1}\cos(n_{x,1}t+ \phi_{x,1})+ \cdots + A_{x,i}\cos(n_{x,i}t+ \phi_{x,i})$$
$$y_{t}=A_{y,1}\cos(n_{y,1}t+ \phi_{y,1})+ \cdots + A_{y,i}\cos(n_{y,i}t+ \phi_{y,i})$$
$$z_{t}=A_{z,1}\cos(n_{z,1}t+ \phi_{z,1})+ \cdots + A_{z,i}\cos(n_{z,i}t+ \phi_{z,i})$$
C. Lamm mostr√≥ que todo nudo se puede representar como un nudo de Fourier-$(1,1,k)$. 


```
import numpy as np
import random
from random import uniform
from sympy import symbols, cos, sin,pi
from sympy.plotting import plot3d_parametric_line

t = symbols('t')
#Coeficientes de x y y son enteros 
ax_1= random.randint(-10,10)
ay_1=random.randint(-10,10)
#Coeficientes  de t de forma entera 
nx_1=random.randint(-10,10)
ny_1=random.randint(-10,10)
phi_1=random.uniform(-10,10)
phi_2=random.uniform(-10,10)
k=2
coeficientes_z=[]
while len(coeficientes_z) <= k-1:
  coeficientes_z.append(random.randint(-10,10))
coeficientes_z_n=[]
while len(coeficientes_z_n) <= k-1:
  coeficientes_z_n.append(random.randint(-10,10))
coeficientes_z_phi=[]
while len(coeficientes_z_phi) <= k-1:
  coeficientes_z_phi.append(uniform(-10,10))

x = ax_1* cos(nx_1*t + phi_1)
y = ay_1 * cos(ny_1*t + phi_2)

z_t=0
for i in range(k):
  z_t= z_t + coeficientes_z[i] * cos (coeficientes_z_n[i] * t + coeficientes_z_phi[i] )

plot3d_parametric_line(x, y, z_t, (t, 0*pi, 2*pi))
```
